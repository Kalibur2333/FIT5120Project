```python
import requests

# Url of the business data
url = "https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/business-establishments-with-address-and-industry-classification/exports/json"

# Calling the data from url
response = requests.get(url)

# Check teh reuest status
if response.status_code == 200:
    # store the data as json
    data = response.json()
    
    # change json to Data Frame
    df1 = pd.DataFrame(data)
    
    # Display DataFrame
    print(df1.head())  # show the first 5 rows of df
else:
    print("Failed to fetch data from the URL.")
```

```python
#change the column containing Year to int.
columns_to_convert = ['census_year']
df1[columns_to_convert] = df1[columns_to_convert].astype(int)
```

```python
# Define connection parameters for RDS
import pymysql
host = 'rds.cnmm0ske6x3q.us-east-1.rds.amazonaws.com'
port = 3306  # Ensure this is a string, not a tuple
database = 'test'
user = 'admin'
password = 'password'
```

```python
# Assume 'done' DataFrame is already defined
try:
    # Connect to the MySQL database
    conn = pymysql.connect(
        host=host,
        port=port,
        database=database,
        user=user,
        password=password
    )
    print("Connected to MySQL database!")

    # Create a cursor object
    cursor = conn.cursor()

    # Check if the table exists

    cursor.execute("SHOW TABLES LIKE 'Location'")
    table_exists = cursor.fetchone()

    if not table_exists:
        # Create the table if it doesn't exist
        create_table_query = """
        CREATE TABLE Location (
            LocationId INT AUTO_INCREMENT PRIMARY KEY,
            LocationName VARCHAR(255),
            LocationAddress VARCHAR(255),
            LocationLatitude FLOAT(18,15),
            LocationLongitude FLOAT(18,15),
            Accepts VARCHAR(255)
        )
        """
        cursor.execute(create_table_query)
        print("Table 'Location' created.")

    # Ensure DataFrame columns match table columns
    done.columns = ['LocationName', 'LocationAddress', 'LocationLatitude', 'LocationLongitude', 'Accepts']

    # Execute SQL query to select all data from the table
    select_query = "SELECT * FROM Location"
    cursor.execute(select_query)

    # Fetch all rows
    rows = cursor.fetchall()

    # Convert the fetched rows to DataFrame
    table_df = pd.DataFrame(rows, columns=[desc[0] for desc in cursor.description])

    # Check if the data in the DataFrame is different from the data in the table
    if not table_df[['LocationName', 'LocationAddress']].equals(done[['LocationName', 'LocationAddress']]):
        
        # Delete existing data from the table
        delete_query = "DELETE FROM Location"
        cursor.execute(delete_query)

        # Insert new data from DataFrame into the table
        insert_query = "INSERT INTO Location (LocationName, LocationAddress, LocationLatitude, LocationLongitude, Accepts) VALUES (%s, %s, %s, %s, %s)"
        values = [tuple(row) for row in done.values]
        cursor.executemany(insert_query, values)

        # Commit changes
        conn.commit()
        print("Database updated successfully.")
    else:
        print("Data in DataFrame is the same as data in the table. No changes made.")

except Exception as e:
    print(f"Unable to connect to MySQL database or perform operation: {e}")

finally:
    # Close the cursor and connection
    if cursor:
        cursor.close()
    if conn:
        print("The connection is closed.")
        conn.close()
```
